```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mariumfaheem666/spam-sms-classification-using-nlp")

print("Path to dataset files:", path)
```

```python
import os
import pandas as pd

# List the contents of the downloaded directory
print(f"Contents of {path}:")
for file in os.listdir(path):
    print(file)
```

From the listed files, it appears `spam.csv` is the data file. Let's load it into a pandas DataFrame.

```python
data_file_path = os.path.join(path, 'Spam_SMS.csv')
df = pd.read_csv(data_file_path)

# Display the first 5 rows of the DataFrame
display(df.head(15))
```

```python
df.shape
```

```python
#text cleaning  with stemming

import re
import nltk
nltk.download('stopwords')
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords

ps=PorterStemmer()

stem_corpus=[]

for i in range(len(df['Message'])):
  review=re.sub('[^a-zA-Z]',' ',df['Message'][i])
  review=review.lower()
  review=review.split()

  revirew=[ps.stem(word) for word in review if word not in set(stopwords.words('english'))]

  review = ' '.join(review)

  stem_corpus.append(review)
```

```python
#Creating BOW model  stemming

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000)
stem_x=cv.fit_transform(stem_corpus).toarray()
```

```python
# converting categorical variables to numerical variables

y=pd.get_dummies(df['Class'])
stem_y=y.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(stem_x,stem_y,test_size=0.20,random_state=0)
```

```python
#model creation
from sklearn.naive_bayes import MultinomialNB
stem_model=MultinomialNB()
stem_model.fit(x_train,y_train)

y_pred=stem_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
stem_accuracy=accuracy_score(y_test,y_pred)
stem_c_matrix=confusion_matrix(y_test,y_pred)
print(stem_c_matrix)
print(stem_accuracy)
```

using lemmatization

```python
#text cleanin with lemmatization

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer=WordNetLemmatizer()

lemma_corpus=[]
for i in range(len(df['Message'])):
  review=re.sub('[^a-zA-Z]',' ',df['Message'][i])
  review=review.lower()
  review=review.split()

  review=[lemmatizer.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
  review=' '.join(review)
  lemma_corpus.append(review)


#BOW for lemma
lemma_cv=CountVectorizer(max_features=500)
lemma_x=lemma_cv.fit_transform(lemma_corpus).toarray()

# converting categorical variables to numerical variables
lemma_y=pd.get_dummies(df['Class'])
lemma_y=lemma_y.iloc[:,1].values

from sklearn.model_selection import train_test_split
lemma_x_train,lemma_x_test,lemma_y_train,lemma_y_test=train_test_split(lemma_x,lemma_y,test_size=0.20,random_state=0)

#model creation
from sklearn.naive_bayes import MultinomialNB
lemma_model=MultinomialNB()
lemma_model.fit(lemma_x_train,lemma_y_train)

y_pred=lemma_model.predict(lemma_x_test)


from sklearn.metrics import confusion_matrix,accuracy_score
lemma_accuracy=accuracy_score(y_test,y_pred)
lemma_c_matrix=confusion_matrix(y_test,y_pred)
print(lemma_c_matrix)
print(lemma_accuracy)
```

using spacy

```python
import spacy
nlp=spacy.load('en_core_web_sm')

spacy_lemma_corpus=[]
for i in range(len(df['Message'])):
  review=re.sub('[^a-zA-Z]',' ',df['Message'][i])

  review=review.lower()

  doc=nlp(review)

  review=[token.lemma_ for token in doc if not token.is_stop]

  review=' '.join(review)

  spacy_lemma_corpus.append(review)

spacy_cv=CountVectorizer(max_features=5000)
spacy_x=spacy_cv.fit_transform(spacy_lemma_corpus).toarray()

spacy_y=pd.get_dummies(df['Class'])
spacy_y=spacy_y.iloc[:,1].values

spacy_x_train,spacy_x_test,spacy_y_train,spacy_y_test=train_test_split(spacy_x,spacy_y,test_size=0.20,random_state=0)

spacy_model=MultinomialNB()
spacy_model.fit(spacy_x_train,spacy_y_train)

y_pred=spacy_model.predict(spacy_x_test)

from sklearn.metrics import confusion_matrix,accuracy_score
spacy_accuracy=accuracy_score(y_test,y_pred)
spacy_c_matrix=confusion_matrix(y_test,y_pred)
print(spacy_c_matrix)
print(spacy_accuracy)
```
